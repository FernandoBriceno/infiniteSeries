# Series de Funciones

Si extendemos lo comprendido sobre series de taal manera que ahora, cada término $u_n$ de nuestra serie sea una función de alguna variable x, $u_n=u_n(x)$.de esta manera las sumas parciales se convierten en funciones de la variable $x$

$s_n(n) = u_1(x)+u_2(x)+...+u_n(x)$

de igual manera la suma de las series, se define como el límite de las sumas parciales

$\sum_{n=1}^{\infty }u_n(x)= S(x) = \lim_{n \to \infty }s_n(x)$

Hasta ahora nos habíamos enfocado en las sumas parciales, las cuales dependen de n, pero ahora, podemos tomarlas como dependientes de $x$.

## Convergencia uniforme

Está relacionado con la tasa de convergencia de una serie de funciones en un intervalo de $x\in [a,b]$. Una serie de funciones continuas convergen en una sola función, también continua, y existen valores límite de la suma para cualquier valor de $x$ dentro del intervalo. 

```python
from IPython import display

print("Weber, Arfken, Esential Mathematical Methods for Physicist, 5th. p. 277")
display.Image("convergencia.PNG")
```

Si para cada $\varepsilon >0$ pequeño, existe un numero $N$, independiente de $x$ en el intervalo $a\leq x\leq b$ tal que 
>$\left | S(x)-s_n(x) \right |<\varepsilon,\, \, para \,todo \, \, n\geq N $

la serie se define como uniformemente convergente en el intervalo $[a,b]$

Visto de otro modo, no importa cuán pequeño se tome $\varepsilon$, siempre podemos elegir un $n$ lo suficientemente grande de tal manera que $\left | S(x)-s_n(x) \right |$ es menor que $\varepsilon$ para todas las $x$ dentro del intervalo
