# Series de Funciones

Si extendemos lo comprendido sobre series de taal manera que ahora, cada término $u_n$ de nuestra serie sea una función de alguna variable x, $u_n=u_n(x)$.de esta manera las sumas parciales se convierten en funciones de la variable $x$

$s_n(n) = u_1(x)+u_2(x)+...+u_n(x)$

de igual manera la suma de las series, se define como el límite de las sumas parciales

$\sum_{n=1}^{\infty }u_n(x)= S(x) = \lim_{n \to \infty }s_n(x)$

Hasta ahora nos habíamos enfocado en las sumas parciales, las cuales dependen de n, pero ahora, podemos tomarlas como dependientes de $x$.

## Convergencia uniforme

Está relacionado con la tasa de convergencia de una serie de funciones en un intervalo de $x\in [a,b]$. Una serie de funciones continuas convergen en una sola función, también continua, y existen valores límite de la suma para cualquier valor de $x$ dentro del intervalo. 

```python
from IPython import display

print("Weber, Arfken, Esential Mathematical Methods for Physicist, 5th. p. 277")
display.Image("IMG/convergencia.PNG")
```

Si para cada $\varepsilon >0$ pequeño, existe un numero $N$, independiente de $x$ en el intervalo $a\leq x\leq b$ tal que 
>$\left | S(x)-s_n(x) \right |<\varepsilon,\, \, para \,todo \, \, n\geq N $

la serie se define como uniformemente convergente en el intervalo $[a,b]$

Visto de otro modo, no importa cuán pequeño se tome $\varepsilon$, siempre podemos elegir un $n$ lo suficientemente grande de tal manera que $\left | S(x)-s_n(x) \right |$ es menor que $\varepsilon$ para todas las $x$ dentro del intervalo

```python
print("by https://www.youtube.com/watch?v=B9TmoYdaMe0")
display.Image("IMG/familias_convergentes.PNG")
```

Tomemos como ejemplo la familia de funciones $\left \{ f_n(x) \right \}= \left \{ \frac{sin(nx)}{n}+1 \,\,\forall n\in \mathbb{N} \right \}$

```python
from math import *
import sympy as sp
from sympy.plotting import plot

x = sp.symbols('x')
f = 1
e_sup = f+0.25
e_inf = f-0.25
f1 = sp.sin(x)+1
f2 = sp.sin(2*x)/2+1
f4 = sp.sin(4*x)/4+1
f8 = sp.sin(8*x)/8+1
f9 = sp.sin(16*x)/16+1


g=plot(f,  f1,f2, f4, f8,f9,e_sup,e_inf,(x,0,2*sp.pi), title='Familia de funciones', show=False)
g[0].line_color ='r'
g[1].line_color ='k'
g[2].line_color ='k'
g[3].line_color ='c'
g[4].line_color ='c'
g[5].line_color ='c'
g[6].line_color ='b'
g[7].line_color ='b'


g.show()



```

Podemos observar que para un $\varepsilon=0.25$ hay una subfamilia de funciones que se acercan a $F(x)=1$ y pueden considerarse convergentes uniformemente, pues siempre se podrá encontrar un n lo suficientemente grande que hará que el valor absoluto de la diferencia entre los máximos de la familia y la función sean menor que $\varepsilon$

Por ejemplo $S_n= \sum_{n=0}^{N}(-1)^nx^{2n} $ es la sumatoria asociada a $f(x) = \frac{1}{1+x^{2}}$ ya que

## $\sum_{n=0}^{\infty }(-1)^nx^{2i} =\frac{1}{1+x^{2}} = S(x) = f(x)$ 

podemos encontrar que existe una subfamilia de funciones para $n \to \infty $ que cumplen el criterio de convergencia uniforme, por ejemplo, si graficamos algunas familias de funciones dentro de esta subfamilia, podemos ver que para el intervalo $-0.5 \leq x \leq 0.5$  son continuas y parecen cumplir $\left | S(x)-s_n(x) \right |<\varepsilon$

```python
x = sp.symbols('x')

def sumatoria(x,n):
    f = 0
    for i in range(0,n):
        f += (-1)**i*x**(2*i)
    return f

s6 = sumatoria(x,6)
s10 = sumatoria(x,10)
s50= sumatoria(x,50)
f= 1/(1+(x**2))

g=plot(f, s6, s10,s50,(x,-1,1), title='Familia de funciones Fn(x) para F(x) = 1/(1+x^2)', show=False)

g[0].line_color ='r'
g[1].line_color ='g'
g[2].line_color ='y'
g[3].line_color ='c'
g.show()

```

Podemos entonces definir de una nueva manera la convergencia. 
>Se dice que una sucesión de funciones $f_n(x)$ **converge uniformemente** a $f(x)$ si la distancia más grande entre la serie de funciones y la función, cuando $n \to \infty$, es igual a cero:
>>$\lim_{ n \to \infty } sup\left \{ \left | f_n(x) - f(x) \right | \right \}=0$

Si encambio tenemos familias de funciones que dentro de un intervalo determinado cerrado, no convergen al mismo punto para todos los valores de x, entonces diremos que la familia es **no uniformemente convergente**. veamos un ejemplo:

$\left \{ f_n(x) \right \}= \left \{ x^n, \,\,\forall n\in \mathbb{N} \right \}\rightarrow f(x)= \left\{\begin{matrix}
0 & si & 0\leq x\leq 1\\ 
1& si & x=1
\end{matrix}\right.$

vemos entonces que hay una discontinuidad en el límite, para todo el intervalo $x\in [0,1]$


```python
x = sp.symbols('x')

f0 = x**0
f1 = x**1
f2 = x**2
f4 = x**4
f8 = x**8

g=plot(f0, f1, f2,f4,f8,(x,0,1), title='Familia de funciones Fn(x) para F(x) = 1/(1+x^2)', show=False)

g[0].line_color ='r'
g[1].line_color ='g'
g[2].line_color ='y'
g[3].line_color ='c'
g[4].line_color ='b'
g.show()
```

Es facil observar que en valores alejados de $x=1$ las funciones tienden a acercarse a cero, pero cuando es muy próximo a $x=1$ tienden a acercarse a $f_n(x)\to 1$

Esto podemos observarlo con el ejemplo ya mencionado para $f(x) = \frac{1}{1+x^{2}}$, si ahora observamos a las familias pero en un intervalo más cercano a 1, podemos ver que antes de $x=1$ las $f_n(x)$ tienden a acercarse a $f(x)$, pero a medida que se acercan a $x=1$ las funciones $f_n(x)\to -\infty$. Veamos:

```python
x = sp.symbols('x')

def sumatoria(x,n):
    f = 0
    for i in range(0,n):
        f += (-1)**i*x**(2*i)
    return f

s6 = sumatoria(x,6)
s10 = sumatoria(x,10)
s50= sumatoria(x,50)
f= 1/(1+(x**2))

g=plot(f, s6, s10,s50,(x,0.5,1.001), title='Familia de funciones Fn(x) para F(x) = 1/(1+x^2)', show=False)

g[0].line_color ='r'
g[1].line_color ='g'
g[2].line_color ='y'
g[3].line_color ='c'
g.show()
```

## Criterio M(mayorante) de Weierstrass 

Es un test que nos permite comprobar **convergencia uniforme**. Si podemos construir una serie de numeros $\sum_{1}^{\infty }M_i$, la cual es convergente, y en la que $M_i \geq \left | u_i(x) \right |$ para  todo $x$ en el intervalo $[a,b]$, entonces nuestras series $u_i(x)$ también serán **uniformemente convergentes** en $[a,b]$

$\exists \sum_{1}^{\infty }  M_i, \: donde\: M_i \geq \left | u_i(x) \right |, \forall i\in \mathbb{N}\Rightarrow   \sum_{1}^{\infty }  u_i\: converge\: absolutamente$

considere $M_i=\frac{1}{i^2}$ y $u_i =\frac{1}{i(i+1)}$

```python
import matplotlib.pyplot as plt
import numpy as np

def M(i):
    M_i = 0
    for j in range(1,i+1):
        m =1/(i**2)
        M_i += m
    return M_i

def U(i):
    U_i = 0
    for j in range(1,i+1):
        u = 1/(i*(i+1))
        U_i += np.abs(u)
    return U_i

i = []
for j in range(1,11):
    i.append(j)
M_i = []
U_i = []
for j in i:
    M_i.append(M(j))
    U_i.append(U(j))

'''Sx = 1
xf = np.linspace(0,10, num=1000)
yf = np.linspace(Sx,Sx, num=1000)'''

plt.figure(num=0, dpi=70)
plt.scatter(i,M_i)
plt.scatter(i,U_i)
#plt.scatter(xf,yf)
plt.xlabel("i-esimo término")
plt.ylabel(" M_i ,  u_i ")
plt.title("")
plt.show
```
